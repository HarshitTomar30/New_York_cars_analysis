# Visualisation

```python
# Bar plot
plt.title('Type description')
df['new&used'].value_counts().plot(kind='bar')
plt.xlabel('new&used')
plt.ylabel('money')

plt.show()
```

```python
# Histogram for year and respective counts
sns.histplot(data=df, x='Year', bins=10,color='red')
plt.xlabel('year')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(data=df, x='Drivetrain', kde=True,color='green')
plt.xlabel('Drivetrain')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=( 20,10))
sns.barplot(data=df, x='Drivetrain', y='Year', hue='Year')
plt.xlabel('Drivetrain')
plt.ylabel('Year')
plt.show()
```

# Visualisation pairwise relationship between variables

```python
# heatmap
corr = df[['Year', 'money', 'Mileage']].corr()
sns.heatmap(corr, annot=True, cmap='cividis')
plt.show()
```

```python
# Pair plot of numeric values
sns.pairplot(data=df[['Year', 'money', 'Mileage']])
plt.show()
```

```python
# line plot of year vs derivetrain
sns.lineplot(data =df, x='Year',y='Exterior',hue='Drivetrain')
plt.xlabel('Year')
plt.ylabel('Drivetrain')
plt.show()
```

```python
# Scatter plot of year verse brand
plt.figure(figsize=(15,10))
sns.scatterplot(data=df, x='Year', y='brand', hue='Drivetrain', palette='cool')
plt.xlabel('Year')
plt.ylabel('brand')
plt.show()
```

```python
df['brand'].value_counts()
```

```python
#voilin plot for o brand verse year
plt.figure(figsize=(15,8))
sns.violinplot(data=df, x='brand', y='Year')
plt.xticks(rotation=60)
plt.xlabel('Brand')
plt.ylabel('Year')
plt.show()
```
